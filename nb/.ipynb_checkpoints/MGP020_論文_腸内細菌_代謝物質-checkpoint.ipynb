## filtering

```python
ge_0001= ge.loc[ge.mean(axis=1) >= 0.001]
ra.drop(ra.loc[ra.index.str.contains("\n")].index, inplace=True)
ra.drop(ra.loc[ra.index.str.startswith("XA")].index, inplace=True)
ra.drop(ra.loc[ra.index.str.startswith("XC")].index, inplace=True)
ra_25 = ra.loc[((ra > 0).sum(axis=1) / len(ra.columns)) >= 0.25]
```

## 差分算出

```python
def get_samesubject_sampleids(mdf, mkey, mvals, msk="subject_id", axis=1, return_subjectid=False):
    def query_col(df, key, val):
        return df.loc[:, df.loc[key, :] == val]
    m = "metadata table is not unique by -mk and -msk"
    subject_ids = sorted(set(mdf.loc[msk, :]))
    sample_ids = []
    for mval in mvals:
        _sample_ids = []
        for subject_id in subject_ids:
            mdf_ext_step1 = query_col(mdf, msk, subject_id)
            mdf_ext_step2 = query_col(mdf_ext_step1, mkey, mval)
            if mdf_ext_step2.shape[1] != 1:
                raise ValueError(m)
            _sample_ids.append(mdf_ext_step2.columns[0])
        sample_ids.append(_sample_ids)

    if return_subjectid and axis == 1:
        return sample_ids, subject_ids
    elif not return_subjectid and axis == 1:
        return sample_ids
    elif return_subjectid and axis == 0:
        return [list(x) for x in zip(*sample_ids)], subject_ids
    elif not return_subjectid and axis == 0:
        return [list(x) for x in zip(*sample_ids)]
```

```python
ge_diff = pd.DataFrame()
ra_diff = pd.DataFrame()
co_diff = pd.DataFrame()
nm_diff = pd.DataFrame()
me_diff = pd.DataFrame()

for mv1, mv2 in zip(["0w"], ["4w"]):
    (sid_ctrls, sid_tests), subject_ids = get_samesubject_sampleids(me, "timepoint", [mv1, mv2], msk="subject_id", return_subjectid=True)
    for sid_ctrl, sid_test, subject_id in zip(sid_ctrls, sid_tests, subject_ids):
        new_sid = f"{subject_id}_{mv2}{mv1}"
        ge_diff[new_sid] = ge_0001[sid_test] - ge_0001[sid_ctrl]
        ra_diff[new_sid] = ra_25[sid_test] - ra_25[sid_ctrl]
        co_diff[new_sid] = co[sid_test] - co[sid_ctrl]
        nm_diff[new_sid] = nm[sid_test] - nm[sid_ctrl]

        for ind in me.index:
            val_ctrl = me.loc[ind, sid_ctrl]
            val_test = me.loc[ind, sid_test]
            if val_ctrl == val_test:
                me_diff.loc[ind, new_sid] = val_ctrl
            else:
                me_diff.loc[ind, new_sid] = f"{val_test}{val_ctrl}"
```

```python
htest_ge_diff = pd.DataFrame(columns=["mw_p", "mw_q", "C_median", "C_IQR", "T_median", "T_IQR"])

samples_cont = me_diff.loc[:, me_diff.loc["group"]=="C"].columns.tolist()
samples_test = me_diff.loc[:, me_diff.loc["group"]=="T"].columns.tolist()

for ind in ge_diff.index:
    vals_cont = ge_diff.loc[ind, samples_cont].values
    vals_test = ge_diff.loc[ind, samples_test].values
    U1, p = st.mannwhitneyu(vals_test, vals_cont, method="asymptotic")
#     n1, n2 = len(vals_test), len(vals_cont)
#     U2 = n1 * n2 - U1
#     z = ((U1 + 0.5) - ((U1 + U2) / 2)) / np.sqrt(n1*n2 * (n1 + n2 + 1)/12)
#     r = z/np.sqrt(n1 + n2)
#     ps = U2/(n1 * n2)
    htest_ge_diff.loc[ind, "mw_p"] = p
#     htest_ge_diff.loc[ind, "mw_r"] = r
#     htest_ge_diff.loc[ind, "mw_ps"] = ps
    htest_ge_diff.loc[ind, "C_median"] = np.median(vals_cont)
    htest_ge_diff.loc[ind, "C_IQR"] = np.percentile(vals_cont, 75) - np.percentile(vals_cont, 25)
    htest_ge_diff.loc[ind, "T_median"] = np.median(vals_test)
    htest_ge_diff.loc[ind, "T_IQR"] = np.percentile(vals_test, 75) - np.percentile(vals_test, 25)
    
htest_ge_diff["mw_q"] = multi.multipletests(htest_ge_diff["mw_p"], alpha=0.05, method='fdr_bh')[1]
```

### 前後比較

```python
htest_ge_C = pd.DataFrame(columns=["wilc_p", "wilc_r", "C_median", "C_IQR", "T_median", "T_IQR"])

n = len(ge_C.columns)

for ind in ge_C.index:
    vals_cont = ge_C.loc[ind, me_C.loc["timepoint"]=="0w"].values
    vals_test = ge_C.loc[ind, me_C.loc["timepoint"]=="4w"].values
    T, p = st.wilcoxon(vals_test, vals_cont, mode="approx")
#     z = st.norm.ppf(1-(p/2))
    z = np.abs((T - (n*2*((n*2)+1))/4))/np.sqrt(n*2*((n*2)+1)*((n*4)+1)/4)
    r = z/np.sqrt(n * 2)
    htest_ge_C.loc[ind, "wilc_p"] = p
    htest_ge_C.loc[ind, "wilc_r"] = r
    htest_ge_C.loc[ind, "C_median"] = np.median(vals_cont)
    htest_ge_C.loc[ind, "C_IQR"] = np.percentile(vals_cont, 75) - np.percentile(vals_cont, 25)
    htest_ge_C.loc[ind, "T_median"] = np.median(vals_test)
    htest_ge_C.loc[ind, "T_IQR"] = np.percentile(vals_test, 75) - np.percentile(vals_test, 25)
```

```python
htest_ge_T = pd.DataFrame(columns=["wilc_p", "wilc_r", "C_median", "C_IQR", "T_median", "T_IQR"])

n = len(ge_T.columns)

for ind in ge_T.index:
    vals_cont = ge_T.loc[ind, me_T.loc["timepoint"]=="0w"].values
    vals_test = ge_T.loc[ind, me_T.loc["timepoint"]=="4w"].values
    if set(vals_cont) == set(vals_test):
        z, p = 0, 1
    else:
        p = st.wilcoxon(vals_test, vals_cont, mode="approx")[1]
        z = st.norm.ppf(1-(p/2))
        r = z/np.sqrt(n * 2)
    htest_ge_T.loc[ind, "wilc_p"] = p
    htest_ge_T.loc[ind, "wilc_r"] = r
    htest_ge_T.loc[ind, "C_median"] = np.median(vals_cont)
    htest_ge_T.loc[ind, "C_IQR"] = np.percentile(vals_cont, 75) - np.percentile(vals_cont, 25)
    htest_ge_T.loc[ind, "T_median"] = np.median(vals_test)
    htest_ge_T.loc[ind, "T_IQR"] = np.percentile(vals_test, 75) - np.percentile(vals_test, 25)
```

```python
htest_ra_C = pd.DataFrame(columns=["wilc_p", "wilc_r", "C_median", "C_IQR", "T_median", "T_IQR"])

n = len(ra_C.columns)

for ind in ra_C.index:
    vals_cont = ra_C.loc[ind, me_C.loc["timepoint"]=="0w"].values
    vals_test = ra_C.loc[ind, me_C.loc["timepoint"]=="4w"].values
    p = st.wilcoxon(vals_test, vals_cont, mode="approx")[1]
    z = st.norm.ppf(1-(p/2))
    r = z/np.sqrt(n * 2)
    htest_ra_C.loc[ind, "wilc_p"] = p
    htest_ra_C.loc[ind, "wilc_r"] = r
    htest_ra_C.loc[ind, "C_median"] = np.median(vals_cont)
    htest_ra_C.loc[ind, "C_IQR"] = np.percentile(vals_cont, 75) - np.percentile(vals_cont, 25)
    htest_ra_C.loc[ind, "T_median"] = np.median(vals_test)
    htest_ra_C.loc[ind, "T_IQR"] = np.percentile(vals_test, 75) - np.percentile(vals_test, 25)
```

```python
htest_ra_T = pd.DataFrame(columns=["wilc_p", "wilc_r", "C_median", "C_IQR", "T_median", "T_IQR"])

n = len(ra_T.columns)

for ind in ra_T.index:
    vals_cont = ra_T.loc[ind, me_T.loc["timepoint"]=="0w"].values
    vals_test = ra_T.loc[ind, me_T.loc["timepoint"]=="4w"].values
    if set(vals_cont) == set(vals_test):
        z, p = 0, 1
    else:
        p = st.wilcoxon(vals_test, vals_cont, mode="approx")[1]
        z = st.norm.ppf(1-(p/2))
        r = z/np.sqrt(n * 2)
    htest_ra_T.loc[ind, "wilc_p"] = p
    htest_ra_T.loc[ind, "wilc_r"] = r
    htest_ra_T.loc[ind, "C_median"] = np.median(vals_cont)
    htest_ra_T.loc[ind, "C_IQR"] = np.percentile(vals_cont, 75) - np.percentile(vals_cont, 25)
    htest_ra_T.loc[ind, "T_median"] = np.median(vals_test)
    htest_ra_T.loc[ind, "T_IQR"] = np.percentile(vals_test, 75) - np.percentile(vals_test, 25)
```

```python
def draw_pp_plot(htest_tp, htest_diff, path_out_pdf=None, labeled=False):
    x_ss = htest_tp.loc[(htest_tp["wilc_p"] < 0.05)&(htest_diff["mw_p"] < 0.05)&(htest_tp["wilc_pGL"] == "G"), "wilc_p"]
    y_ss = htest_diff.loc[(htest_tp["wilc_p"] < 0.05)&(htest_diff["mw_p"] < 0.05)&(htest_tp["wilc_pGL"] == "G"), "mw_p"]
    x_s = htest_tp.drop(x_ss.index).loc[(htest_tp["wilc_p"] < 0.05)&(htest_diff["mw_p"] < 0.05)&(htest_tp["wilc_pGL"] == "L"), "wilc_p"]
    y_s = htest_diff.drop(x_ss.index).loc[(htest_tp["wilc_p"] < 0.05)&(htest_diff["mw_p"] < 0.05)&(htest_tp["wilc_pGL"] == "L"), "mw_p"]
    x_ns = htest_tp.drop(x_ss.index.tolist() + x_s.index.tolist())["wilc_p"]
    y_ns = htest_diff.drop(y_ss.index.tolist() + y_s.index.tolist())["mw_p"]
    
    x_s_log10 = [-1 * np.log10(val) for val in x_s]
    y_s_log10 = [-1 * np.log10(val) for val in y_s]
    x_ss_log10 = [-1 * np.log10(val) for val in x_ss]
    y_ss_log10 = [-1 * np.log10(val) for val in y_ss]
    x_ns_log10 = [-1 * np.log10(val) for val in x_ns]
    y_ns_log10 = [-1 * np.log10(val) for val in y_ns]
    
    # 正方形にする
    plt.figure(figsize=(4, 4))
    
    # main_scatter
    plt.scatter(x_ns_log10, y_ns_log10, s=16, alpha=0.5, c="gray")
    plt.scatter(x_s_log10, y_s_log10, s=16, alpha=0.5, c="blue")
    plt.scatter(x_ss_log10, y_ss_log10, s=16, alpha=0.5, c="red")
    
#     if labeled:
#         for _x, _y, _text in zip(x_ss, y_ss, x_ss.keys()):
#             _x_log10 = -1 * np.log10(_x)
#             _y_log10 = -1 * np.log10(_y)
#             plt.text(_x_log10, _y_log10, _text, size=6)

#         for _x, _y, _text in zip(x_s, y_s, x_s.keys()):
#             _x_log10 = -1 * np.log10(_x)
#             _y_log10 = -1 * np.log10(_y)
#             plt.text(_x_log10, _y_log10, _text, size=6)
    
    # 表示の大きい方に合わせる
    xmax, ymax = plt.xlim()[1], plt.ylim()[1]
    xymax = max([xmax, ymax])
    plt.xlim([plt.xlim()[0], xymax])
    plt.ylim([plt.ylim()[0], xymax])
    
    # 0.05のlineに線を引く
    cutoff = -1 * np.log10(0.05)
    plt.hlines(y=cutoff, xmin=plt.xlim()[0], xmax=xymax, colors='gray', linestyles="dashed")
    plt.vlines(x=cutoff, ymin=plt.ylim()[0], ymax=xymax, colors='gray', linestyles="dashed")
    
    # 右と上非表示
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    
#     if path_out_pdf is not None:
#         plt.savefig(path_out_pdf, bbox_inches="tight")
        
    plt.show()
```

```python
draw_pp_plot(htest_ge_C, htest_ge_diff, "pp_plot.ge.C.pdf")
draw_pp_plot(htest_ge_T, htest_ge_diff, "pp_plot.ge.T.pdf")
draw_pp_plot(htest_ra_C, htest_ra_diff, "pp_plot.ra.C.pdf")
draw_pp_plot(htest_ra_T, htest_ra_diff, "pp_plot.ra.T.pdf")
```
