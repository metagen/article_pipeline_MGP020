```python
import numpy as np
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
nm_C = pd.read_table("../data_inter/nm2.C.tsv", index_col=0)
nm_T = pd.read_table("../data_inter/nm2.T.tsv", index_col=0)
me_C = pd.read_table("../data_inter/me2.C.tsv", index_col=0)
me_T = pd.read_table("../data_inter/me2.T.tsv", index_col=0)

nm_diff_2w0w = pd.read_table("../data_inter/nm2.diff.2w.0w.tsv", index_col=0)
nm_diff_4w0w = pd.read_table("../data_inter/nm2.diff.4w.0w.tsv", index_col=0)
me_diff_2w0w = pd.read_table("../data_inter/me2.diff.2w.0w.tsv", index_col=0)
me_diff_4w0w = pd.read_table("../data_inter/me2.diff.4w.0w.tsv", index_col=0)
```

```python
df_cat_diff_2w0w = pd.concat([nm_diff_2w0w, me_diff_2w0w]).T.melt(id_vars=me_diff_2w0w.index).query("variable == 'IgA'")
df_cat_diff_4w0w = pd.concat([nm_diff_4w0w, me_diff_4w0w]).T.melt(id_vars=me_diff_4w0w.index).query("variable == 'IgA'")

df_cat_diff_2w0w_C = df_cat_diff_2w0w.query("group == 'C'")
df_cat_diff_2w0w_T = df_cat_diff_2w0w.query("group == 'T'")
df_cat_diff_4w0w_C = df_cat_diff_4w0w.query("group == 'C'")
df_cat_diff_4w0w_T = df_cat_diff_4w0w.query("group == 'T'")
```

## 棒グラフ

```python
plt.rcParams["font.family"] = "Arial"
```

```python
!mkdir -p ../data_output/IgA_barplot
```

```python
plt.figure(figsize=(8, 5))
vals = df_cat_diff_2w0w_C.sort_values("value", ascending=False)
clrs = ['red' if (x > 0) else 'blue' for x in vals["value"] ]
sns.barplot(x="subject_id", y="value", data=df_cat_diff_2w0w_C, order=vals["subject_id"], palette=clrs)
plt.ylabel("Δfecal lgA (ug/mg dry feces)", size=16)
plt.xlabel("")
plt.title("2w - 0w (CF)", size=20)
plt.xticks(rotation=90)
plt.savefig("../data_output/IgA_barplot/IgA.barplot.C.2w0w.pdf", bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
vals = df_cat_diff_4w0w_C.sort_values("value", ascending=False)
clrs = ['red' if (x > 0) else 'blue' for x in vals["value"] ]
sns.barplot(x="subject_id", y="value", data=df_cat_diff_4w0w_C, order=vals["subject_id"], palette=clrs)
plt.ylabel("Δfecal lgA (ug/mg dry feces)", size=16)
plt.xlabel("")
plt.title("4w - 0w (CF)", size=20)
plt.xticks(rotation=90)
plt.savefig("../data_output/IgA_barplot/IgA.barplot.C.4w0w.pdf", bbox_inches="tight")
plt.show()
```

## カウント

```python
df_count = pd.DataFrame()

df_count.loc["2w - 0w (CF)", "Increase"] = df_cat_diff_2w0w_C.query("value > 0").shape[0]
df_count.loc["2w - 0w (CF)", "Decrease"] = df_cat_diff_2w0w_C.query("value < 0").shape[0]
df_count.loc["4w - 0w (CF)", "Increase"] = df_cat_diff_4w0w_C.query("value > 0").shape[0]
df_count.loc["4w - 0w (CF)", "Decrease"] = df_cat_diff_4w0w_C.query("value < 0").shape[0]

df_count.loc["2w - 0w (FG)", "Increase"] = df_cat_diff_2w0w_T.query("value > 0").shape[0]
df_count.loc["2w - 0w (FG)", "Decrease"] = df_cat_diff_2w0w_T.query("value < 0").shape[0]
df_count.loc["4w - 0w (FG)", "Increase"] = df_cat_diff_4w0w_T.query("value > 0").shape[0]
df_count.loc["4w - 0w (FG)", "Decrease"] = df_cat_diff_4w0w_T.query("value < 0").shape[0]
```
