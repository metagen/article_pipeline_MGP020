```python
import numpy as np
import pandas as pd
import scipy.stats as st

from IPython.display import display
```

# Table 2

```python
me = pd.read_csv("../data_input/me.tsv", sep="\t", header=0, index_col=0)
nm = pd.read_csv("../data_input/nm.tsv", sep="\t", header=0, index_col=0)
```

```python
nm
```

```python
df = pd.concat([me, nm])
```

```python
df_C_0w = df.T.query("group_timepoint == 'C_0w'").T
df_T_0w = df.T.query("group_timepoint == 'T_0w'").T
df_C_2w = df.T.query("group_timepoint == 'C_2w'").T
df_T_2w = df.T.query("group_timepoint == 'T_2w'").T
df_C_4w = df.T.query("group_timepoint == 'C_4w'").T
df_T_4w = df.T.query("group_timepoint == 'T_4w'").T
df_All = df.T.query("timepoint == '0w'").T
```

```python
df_out2 = pd.DataFrame(
    columns=["CF", "FG", "All"],
    index=["sex", "age", "BMI", "排便回数", "量"]
)
```

```python
for ind in df_out2.index:
    if ind == "sex":
        count_C = df_C.loc['sex'].value_counts()
        count_T = df_T.loc['sex'].value_counts()
        count_All = df_All.loc['sex'].value_counts()
        df_out2.loc[ind, "CF"] = f"{count_C.male}, {count_C.female}"
        df_out2.loc[ind, "FG"] = f"{count_T.male}, {count_T.female}"
        df_out2.loc[ind, "All"] = f"{count_All.male}, {count_All.female}"
        df_out2.loc[ind, "pval"] = "-"
    else:
        mean_C = df_C.loc[ind].astype(float).mean()
        mean_T = df_T.loc[ind].astype(float).mean()
        mean_All = df.loc[ind].astype(float).mean()
        std_C = df_C.loc[ind].astype(float).std()
        std_T = df_T.loc[ind].astype(float).std()
        std_All = df_All.loc[ind].astype(float).std()
        p = st.ttest_ind(df_C.loc[ind].astype(float), df_T.loc[ind].astype(float), equal_var=False).pvalue
        df_out2.loc[ind, "CF"] = f"{round(mean_C, 1)} ± {round(std_C, 1)}"
        df_out2.loc[ind, "FG"] = f"{round(mean_T, 1)} ± {round(std_T, 1)}"
        df_out2.loc[ind, "All"] = f"{round(mean_All, 1)} ± {round(std_All, 1)}"
        df_out2.loc[ind, "pval"] = round(p, 2)
```

# Table 3, 4

```python
me = pd.read_csv("../data_input/me2.tsv", sep="\t", header=0, index_col=0)
nm = pd.read_csv("../data_input/nm2.tsv", sep="\t", header=0, index_col=0)

df = pd.concat([me, nm])

df_C_0w = df.T.query("group_timepoint == 'C_0w'").T
df_C_2w = df.T.query("group_timepoint == 'C_2w'").T
df_C_4w = df.T.query("group_timepoint == 'C_4w'").T
df_T_2w = df.T.query("group_timepoint == 'T_2w'").T
df_T_0w = df.T.query("group_timepoint == 'T_0w'").T
df_T_4w = df.T.query("group_timepoint == 'T_4w'").T

pvals_C_2w = pd.read_csv("../data_output/htest_tp/nm2.htest.0w.2w.C.tsv", sep="\t", header=0, index_col=0)
pvals_C_4w = pd.read_csv("../data_output/htest_tp/nm2.htest.0w.4w.C.tsv", sep="\t", header=0, index_col=0)
pvals_T_2w = pd.read_csv("../data_output/htest_tp/nm2.htest.0w.2w.T.tsv", sep="\t", header=0, index_col=0)
pvals_T_4w = pd.read_csv("../data_output/htest_tp/nm2.htest.0w.4w.T.tsv", sep="\t", header=0, index_col=0)
```

```python
display(df_C_0w.head())
display(df_C_2w.head())
display(df_C_4w.head())
display(df_T_0w.head())
display(df_T_2w.head())
display(df_T_4w.head())
```

```python
display(df_C_0w.tail())
display(df_C_2w.tail())
display(df_C_4w.tail())
display(df_T_0w.tail())
display(df_T_2w.tail())
display(df_T_4w.tail())
```

```python
df_out3 = pd.DataFrame(
    index=pvals_C_2w.index
)

for ind in df_out3.index:
    mean_C_0w = df_C_0w.loc[ind].astype(float).mean()
    std_C_0w = df_C_0w.loc[ind].astype(float).std()
    mean_C_2w = df_C_2w.loc[ind].astype(float).mean()
    std_C_2w = df_C_2w.loc[ind].astype(float).std()
    mean_C_4w = df_C_4w.loc[ind].astype(float).mean()
    std_C_4w = df_C_4w.loc[ind].astype(float).std()
    p_C_2w = pvals_C_2w.loc[ind, "wilcoxon"]
    q_C_2w = pvals_C_2w.loc[ind, "wilcoxon_f"]
    p_C_4w = pvals_C_4w.loc[ind, "wilcoxon"]
    q_C_4w = pvals_C_4w.loc[ind, "wilcoxon_f"]
    df_out3.loc[ind, "CF_0w_mean"] = round(mean_C_0w, 2)
    df_out3.loc[ind, "CF_0w_std"] = round(std_C_0w, 2)
    df_out3.loc[ind, "CF_2w_mean"] = round(mean_C_2w, 2)
    df_out3.loc[ind, "CF_2w_std"] = round(std_C_2w, 2)
    df_out3.loc[ind, "CF_pval_2w"] = round(p_C_2w, 3)
    df_out3.loc[ind, "CF_qval_2w"] = round(q_C_2w, 3)
    df_out3.loc[ind, "CF_4w_mean"] = round(mean_C_4w, 2)
    df_out3.loc[ind, "CF_4w_std"] = round(std_C_4w, 2)
    df_out3.loc[ind, "CF_pval_4w"] = round(p_C_4w, 3)
    df_out3.loc[ind, "CF_qval_4w"] = round(q_C_4w, 3)
```

```python
df_out4 = pd.DataFrame(
    index=pvals_T_2w.index
)

for ind in df_out4.index:
    mean_T_0w = df_T_0w.loc[ind].astype(float).mean()
    std_T_0w = df_T_0w.loc[ind].astype(float).std()
    mean_T_2w = df_T_2w.loc[ind].astype(float).mean()
    std_T_2w = df_T_2w.loc[ind].astype(float).std()
    mean_T_4w = df_T_4w.loc[ind].astype(float).mean()
    std_T_4w = df_T_4w.loc[ind].astype(float).std()
    p_T_2w = pvals_T_2w.loc[ind, "wilcoxon"]
    q_T_2w = pvals_T_2w.loc[ind, "wilcoxon_f"]
    p_T_4w = pvals_T_4w.loc[ind, "wilcoxon"]
    q_T_4w = pvals_T_4w.loc[ind, "wilcoxon_f"]
    df_out4.loc[ind, "FG_0w_mean"] = round(mean_T_0w, 2)
    df_out4.loc[ind, "FG_0w_std"] = round(std_T_0w, 2)
    df_out4.loc[ind, "FG_2w_mean"] = round(mean_T_2w, 2)
    df_out4.loc[ind, "FG_2w_std"] = round(std_T_2w, 2)
    df_out4.loc[ind, "FG_pval_2w"] = round(p_T_2w, 3)
    df_out4.loc[ind, "FG_qval_2w"] = round(q_T_2w, 3)
    df_out4.loc[ind, "FG_4w_mean"] = round(mean_T_4w, 2)
    df_out4.loc[ind, "FG_4w_std"] = round(std_T_4w, 2)
    df_out4.loc[ind, "FG_pval_4w"] = round(p_T_4w, 3)
    df_out4.loc[ind, "FG_qval_4w"] = round(q_T_4w, 3)
```
