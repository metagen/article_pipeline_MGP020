## Table 1

```python
df1 = pd.concat([me, nm])
df1_C = df1.T.query("group_timepoint == 'C_0w'")
df1_T = df1.T.query("group_timepoint == 'T_0w'")
```

```python
df_out1 = pd.DataFrame(columns = ["CF", "FG", "p-value"])
```

```python
# Subjects
df_out1.loc["Subjects", "CF"] = df1_C.shape[0]
df_out1.loc["Subjects", "FG"] = df1_T.shape[0]

# Male Female Count
sex_count_C = df1_C["sex"].value_counts()
sex_count_T = df1_T["sex"].value_counts()
df_out1.loc["Sex(male, female)", "CF"] = f"{sex_count_C.male}, {sex_count_C.female}"
df_out1.loc["Sex(male, female)", "FG"] = f"{sex_count_T.male}, {sex_count_T.female}"
df_out1.loc["Sex(male, female)", "p-value"] = "-"

# Age
age_mean_C = round(df1_C["age"].astype(float).mean(), 1)
age_mean_T = round(df1_T["age"].astype(float).mean(), 1)
age_std_C = round(df1_C["age"].astype(float).std(), 1)
age_std_T = round(df1_T["age"].astype(float).std(), 1)
age_pvalue = st.ttest_ind(df1_C["age"].astype(float), df1_T["age"].astype(float), equal_var=False).pvalue

df_out1.loc["Age(years)", "CF"] = f"{age_mean_C}±{age_std_C}"
df_out1.loc["Age(years)", "FG"] = f"{age_mean_T}±{age_std_T}"
df_out1.loc["Age(years)", "p-value"] = round(age_pvalue, 2)

# BMI
age_mean_C = round(df1_C["BMI"].astype(float).mean(), 1)
age_mean_T = round(df1_T["BMI"].astype(float).mean(), 1)
age_std_C = round(df1_C["BMI"].astype(float).std(), 1)
age_std_T = round(df1_T["BMI"].astype(float).std(), 1)
age_pvalue = st.ttest_ind(df1_C["BMI"].astype(float), df1_T["BMI"].astype(float), equal_var=False).pvalue

df_out1.loc["BMI (kg/m)2", "CF"] = f"{age_mean_C}±{age_std_C}"
df_out1.loc["BMI (kg/m)2", "FG"] = f"{age_mean_T}±{age_std_T}"
df_out1.loc["BMI (kg/m)2", "p-value"] = round(age_pvalue, 2)

# Defecation frequency
age_mean_C = round(df1_C["Defecation frequency (bowel movement counts/days)"].astype(float).mean(), 1)
age_mean_T = round(df1_T["Defecation frequency (bowel movement counts/days)"].astype(float).mean(), 1)
age_std_C = round(df1_C["Defecation frequency (bowel movement counts/days)"].astype(float).std(), 1)
age_std_T = round(df1_T["Defecation frequency (bowel movement counts/days)"].astype(float).std(), 1)
age_pvalue = st.ttest_ind(df1_C["Defecation frequency (bowel movement counts/days)"].astype(float), df1_T["Defecation frequency (bowel movement counts/days)"].astype(float), equal_var=False).pvalue

df_out1.loc["Defecation frequency (times/day)", "CF"] = f"{age_mean_C}±{age_std_C}"
df_out1.loc["Defecation frequency (times/day)", "FG"] = f"{age_mean_T}±{age_std_T}"
df_out1.loc["Defecation frequency (times/day)", "p-value"] = round(age_pvalue, 2)

df_out1
```
